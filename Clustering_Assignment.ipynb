#Question 1... What is unsupervised learning in the context of machine learning?


- Unsupervised learning is a type of machine learning where the model is trained without labeled data.
That means the algorithm is not told what the correct answers are — it must discover patterns, structure, or relationships on its own from raw input data.

- Main tasks:
1. Clustering

   - Grouping similar data points.

- Examples:

  - Grouping customers into buying-behavior categories

  - Separating similar images

2. Dimensionality Reduction

- Examples:

- Examples:

4. Association Rule Learning

   - Finding relationships between items.

- Example:

   - Market basket analysis (People who buy bread also buy butter)


#Question 2...  How does K-Means clustering algorithm work?

1. Choose the number of clusters (K)

   - Example: K = 3 → You want 3 clusters.

2. Initialize centroids randomly

   - Pick K random points from the dataset as the starting “centers” (centroids).

   - These centroids represent the center of each cluster.

3. Assign points to the nearest centroid (Cluster assignment step)

    - For every data point:

      - Calculate the distance to all centroids (usually Euclidean distance).

      - Assign each point to the cluster of the nearest centroid.

4. Update the centroid positions (Centroid update step)

   - For each cluster:

      - Recalculate the centroid = mean of all points in that cluster.


# Question 3... Explain the concept of a dendrogram in hierarchical clustering.

- A dendrogram shows:

1. How clusters are formed step by step

    - At the bottom → each individual data point starts as its own cluster.

   - As you move up → clusters start merging based on similarity.

    - At the top → all points merge into one cluster.

2. The distance between clusters

   - The height at which two clusters merge represents how similar (or different) they are.

    - Low merge height = clusters are very similar

    - High merge height = clusters are less similar.

- Types of Hierarchical Clustering Shown in Dendrograms
1. Agglomerative (Bottom-Up)

    - Starts with each point as a cluster

    - Merges clusters step by step

    - Most common (and what dendrograms usually show)

2. Divisive (Top-Down)

   - Starts with one big cluster

   - Splits it repeatedly

#Question 4... What is the main difference between K-Means and Hierarchical Clustering?

- Main Difference
    -  K-Means requires you to choose the number of clusters (K) in advance, but Hierarchical Clustering does NOT — it builds a tree (dendrogram) and you can decide the number of clusters later.

- Key Difference:-

1. Need for number of clusters

    - K-Means: You must specify K before running the algorithm.

   - Hierarchical: No need to predefine K; you cut the dendrogram at any level.

2. How clusters are formed

   - K-Means: Creates clusters by moving centroids and reassigning points (iterative).

   - Hierarchical: Builds a tree by repetitive merging (agglomerative) or splitting (divisive).

3. Output form

    - K-Means: Gives exact K clusters.

    - Hierarchical: Gives a dendrogram, showing how clusters form at every step.


#Question 5... What are the advantages of DBSCAN over K-Means?

- Advantages:-

1. DBSCAN does not require specifying the number of clusters(K).
   - K-Means needs you to choose K beforehand, which is often difficult.
2. DBSCAN can find clusters of any shape.
   - K-Means can only find spherical (round) clusters.
3. DBSCAN handles noise and outliers very well.

    - It labels isolated points as noise, instead of forcing them into a cluster.
4. DBSCAN can handle clusters of different sizes and densities(better than K-Means)
    - DBSCAN groups points based on density, so size differences matter less.

5. DBSCAN is deterministic.
   - DBSCAN gives same clusters every time (unless distance ties occur).

#Question 9...  How does DBSCAN identify noise points?


- Three types of Points:-
1. Core Point

- A point is a core point if:

    - It has at least minPts points (including itself)

     - Within a distance of eps

2. Border Point

- A point that:

    - Has fewer than minPts within eps  

3. Noise Point (Outlier)

- A point is noise if:

   - It is not a core point      
   - AND not within eps of any core point

#Question 10... Define inertia in the context of K-Means?

- Inertia is the sum of squared distances between each data point and the centroid of the cluster it belongs to.

- What Inertia Represents

    - Lower inertia = tighter, better-defined clusters

    - Higher inertia = points are spread out (poor clustering)

    - K-Means tries to minimize inertia during training.

- Why Inertia Is Important
1. Evaluating cluster quality

   - Lower inertia means:

   - Points are close to their centroids

   - Clusters are compact

2. Choosing the number of clusters (K)

   - Used in the Elbow Method:

    - Plot inertia vs K

    - Choose the “elbow” where inertia starts decreasing slowly

#Question 11... What is the elbow method in K-Means clustering?

- The Elbow Method is a technique used to decide the optimal number of clusters (K) in K-Means clustering.

- How It Works

    - Run K-Means for different values of K (like 1 to 10).

    - For each K, compute inertia (also called WCSS – Within Cluster Sum of Squares).

    - Plot K vs. Inertia.

    - The curve will look like a downward slope — inertia decreases as K increases.

    - At some point, the rate of decrease in inertia slows down sharply.

- It helps you find the best value of K by checking how the inertia (within-cluster sum of squared distances) decreases as K increases.    


#Question 12...  Describe the concept of "density" in DBSCAN?

- Density refers to how closely packed the data points are in a region.
- DBSCAN groups points together only if they are in a region with enough nearby points.

- How DBSCAN Uses Density

 - DBSCAN classifies points based on density:

1. Core Point

    - A point is a core point if:

    - It has at least MinPts neighbors within ε.
    - This means it is in a dense region.

2. Border Point

    - A point that is not dense enough to be a core point but is close to a core point.

3. Noise Point

    - A point that is not in any dense region.

- Categorical data needs special distance metrics or encoding techniques for hierarchical clustering to work correctly.

1. Use a distance metric designed for categorical data
   - Works on:

       - categorical

       - numerical
2. Convert (encode) categorical data into numbers
    - One hot encoding
    - Ordinal Encoding
3. Use specialized clustering algorithms (when categories are many)
   - K-Modes (for purely categorical data)

   - K-Prototypes (for mixed numeric + categorical data)

#Question 14...    What does a negative Silhouette Score indicate?

- A negative Silhouette Score indicates bad clustering — meaning the data point is likely assigned to the wrong cluster.

- What Silhouette Score Measures

    - Silhouette Score compares two things for each point:

       - a = how close the point is to points in its own cluster

       - b = how close the point is to points in the nearest other cluster

- What it Indicates

    - The point is likely misclassified

    - The clusters are overlapping

    - The chosen number of clusters (K) may be wrong

     - The clustering algorithm didn’t form clear boundaries       

#Question 15...  Explain the term "linkage criteria" in hierarchical clustering.

- In hierarchical clustering, the term "linkage criteria" refers to the rule used to measure the distance between two clusters when deciding which clusters should be merged next.

- Since hierarchical clustering builds clusters step-by-step, it must always choose which two clusters are closest.

- Types of Linkage Criteria
1. Single Linkage

     - Distance between two clusters = distance between their closest points

2. Complete Linkage

     - Distance between clusters = distance between their farthest points     

3. Average Linkage

    - Distance = average distance between all pairs of points     

4. Ward’s Linkage

    - Merges clusters that result in the minimum increase in total variance

#Question 16...  Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

- K-Means can perform poorly on data with varying cluster sizes or densities because of its built-in assumptions.

1.  K-Means assumes clusters are spherical and equal in size

    - K-Means uses Euclidean distance and tries to form clusters that are:

           - round

           - same size

2. Centroids get pulled toward larger or denser clusters

    - K-Means updates centroids by taking the mean of points.


3. K-Means uses variance as a measure — sensitive to density

    - Sparse clusters have high variance.
    - Dense clusters have low variance.

#Question 17..  What are the core parameters in DBSCAN, and how do they influence clustering?

- DBSCAN has two core parameters that completely control how clustering works:

1. (epsilon)

    - The maximum radius around a point to look for neighbors.

2. MinPts (minimum points)

    - The minimum number of points required inside the ε-radius for a region to be considered dense.

- DBSCAN identifies:

1.  Core points

    - Points with ≥ MinPts neighbors within epsilon

2.  Border points

     - Within epsilon of a core point but with fewer than MinPts neighbors

3. Noise points

    - Points that don’t fit into any dense region  

#Question 18...  How does K-Means++ improve upon standard K-Means initialization?

- K-Means++ improves standard K-Means by choosing better initial centroids, which leads to:

   - faster convergence
   - more stable results
   -  better clustering quality
- Why Standard K-Means Fails Sometimes

- Standard K-Means picks random initial centroids.
    - This can lead to:

        - bad starting points

        - poor cluster separation

        - the algorithm getting stuck in a bad local minimum

        - different results each time   

#Question 19... What is agglomerative clustering?

- Agglomerative clustering is the most common type of hierarchical clustering, and it works by building clusters from the bottom up.

- Agglomerative clustering starts with each data point as its own cluster, and then repeatedly merges the two closest clusters until only one big cluster remains (or until you decide to stop).

- Agglomerative clustering naturally produces a dendrogram — a tree-like diagram showing how clusters merged at each step.

- You can “cut” the dendrogram at any height to choose the number of clusters.

#Question 20... What makes Silhouette Score a better metric than just inertia for model evaluation?

- Silhouette Score is considered better than just using inertia for evaluating clustering because it gives more meaningful, interpretable, and reliable information about the quality of clusters.

1. Inertia only measures compactness — NOT separation

2. Silhouette Score considers BOTH compactness and separation
3. Silhouette Score detects misclassified points
4. Inertia always decreases with more clusters — Silhouette does NOT

5. Silhouette Score works across different algorithms




#Question 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.


```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y_true = make_blobs(
    n_samples=500,
    centers=4,
    cluster_std=1.0,
    random_state=42
)

kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    s=200,
    marker='X'
)

plt.title("K-Means Clustering with 4 Synthetic Blobs")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

#Question 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, y = make_blobs(
    n_samples=500,
    centers=4,
    cluster_std=1.0,
    random_state=42
)

ks = [2, 3, 4, 5]
scores = []

for k in ks:
    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

plt.figure(figsize=(7, 5))
plt.bar(ks, scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()
```

#Question 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(
    n_samples=600,
    centers=3,
    cluster_std=[1.5, 1.5, 1.5],
    random_state=42
)

kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)
centers = kmeans.cluster_centers_

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 500),
    np.linspace(y_min, y_max, 500)
)

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 7))

plt.contourf(xx, yy, Z, alpha=0.25)

plt.scatter(X[:, 0], X[:, 1], c=labels, s=30)

plt.scatter(
    centers[:, 0], centers[:, 1],
    marker='X', s=200, edgecolor='black'
)

plt.title("K-Means Clustering with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

#Question 34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.

#Question 44.  Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

#Question 46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd

data = load_wine()
X = data.data
feature_names = data.feature_names

X_selected = X[:, :4]

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_selected)

df = pd.DataFrame(X_selected, columns=feature_names[:4])
df['Cluster'] = labels

sns.pairplot(df, hue='Cluster', diag_kind='kde', palette='Set2')
plt.suptitle("KMeans Clustering on Wine Dataset (First 4 Features)", y=1.02)
plt.show()
```

#Question 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

X, y = make_blobs(
    n_samples=800,
    centers=3,
    cluster_std=1.2,
    random_state=42
)

noise = np.random.uniform(low=-10, high=10, size=(80, 2))
X_noisy = np.vstack([X, noise])

dbscan = DBSCAN(eps=0.7, min_samples=5)
labels = dbscan.fit_predict(X_noisy)


num_noise = list(labels).count(-1)
num_clusters = len(set(labels)) - (1 if -1 in labels else 0)

print("Number of clusters found:", num_clusters)
print("Number of noise points:", num_noise)
```

#Question 48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

digits = load_digits()
X = digits.data

tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X)

agg = AgglomerativeClustering(n_clusters=10)
labels = agg.fit_predict(X_tsne)

plt.figure(figsize=(10, 7))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=15)
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE 2D)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
